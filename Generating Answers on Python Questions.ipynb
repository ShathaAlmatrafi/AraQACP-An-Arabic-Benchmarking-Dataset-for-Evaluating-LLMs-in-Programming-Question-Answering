```python
ANSWER_SYSTEM_PROMPT = """
You are an expert Python programming instructor.

- Always answer in clear, Modern Standard Arabic (no dialect).
- Keep all Python code, keywords, and identifiers in English.
- Assume the question is about Python and give correct, concise explanations.
- When helpful, include small Python code examples.

There are two learner types:
- Beginner: little or no programming background → explain slowly, avoid heavy jargon, use simple examples.
- Experience in Programming: knows general programming → be a bit more concise, focus on Python-specific details.
"""
```

```python
def generate_answer_ar(question_ar: str, model: str = "gpt-5-nano") -> str:
    if not isinstance(question_ar, str) or question_ar.strip() == "":
        return ""

    user_content = f"""
السؤال في برمجة بايثون:

{question_ar}

اكتب إجابة نموذجية باللغة العربية الفصحى، مناسبة لمتعلم بايثون، ويمكنك استخدام أمثلة كود بايثون إذا لزم الأمر.
"""

    resp = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": ANSWER_SYSTEM_PROMPT},
            {"role": "user", "content": user_content},
        ]
        # no temperature parameter for gpt-5-nano
    )

    return resp.choices[0].message.content.strip()
```

```python
sample_row = df.iloc[0]

print("=== Arabic Question ===")
print(sample_row["question_ar"])

print("\n=== Expert Arabic Answer (translated) ===")
print(sample_row["answer_ar"][:400])

print("\n=== Model Arabic Answer (generated) ===")
print(generate_answer_ar(sample_row["question_ar"])[:600])
```

```python
df["answer_model_ar"] = ""

for idx, row in df.iterrows():
    q_ar = row["question_ar"]

    df.at[idx, "answer_model_ar"] = generate_answer_ar(q_ar)

    if (idx + 1) % 20 == 0:
        print(f"Generated answers for {idx + 1} questions...")
```

```python
output_path = "QACP_Python_QA_ar_with_model_answers.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")
print("Saved to:", output_path)
```
